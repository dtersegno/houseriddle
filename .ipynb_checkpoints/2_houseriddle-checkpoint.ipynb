# Tabular

````python
pd.options.display.max_columns = 30
all_nodes = np.array(nationalities+colors+pets+cigars+house_nos+drinks, dtype = 'str')
````

````python
direct_hints = [ 'English-red',
 'Swedish-dogs',
 'Danish-tea',
 'green-coffee',
 'Pall Mall-birds',
 'yellow-Dunhill',
 '2-milk',
 'Norwegian-0',
 'Blue Masters-beer',
 'German-Prince',
 'Norwegian-blue',
]

neighbor_hints = ['Blendneighborscats', 'horsesneighborsDunhill', 'Blendneighborswater']

direct_hints = list(map(lambda x: x.split('-'), direct_hints))
neighbor_hints = list(map(lambda x: x.split('neighbors'), neighbor_hints))
````

````python
df = pd.DataFrame([[None]*30]*30, columns = all_nodes)
df.index = all_nodes

def sym_check(list_of_paired_nodes, value = 1, df = df):
    index, column = list_of_paired_nodes
    df.loc[index, column] = value
    df.loc[column, index] = value
    return None
````

````python
for pair in direct_hints:
    sym_check(pair)
for pair in neighbor_hints:
    sym_check(pair, value = 0)
df
````

````python
setup_query = """UNWIND {} AS new_node_name
CREATE (node:{} [[thing:new_node_name, lbl:"{}"]]);
"""

def create_node_string(object_list, label):
    query = setup_query.format(object_list, label, label.lower())
    return query.replace('[[','{').replace(']]','}')

def print_node_string(object_list, label):
    print(create_node_string(object_list, label))
    return None
````

````python
print_node_string(nationalities, "Nationality")
print_node_string(colors, "Color")
print_node_string(pets, "Pet")
print_node_string(house_nos, "House")
print_node_string(cigars, "Cigar")
print_node_string(drinks, "Drink")
````

# Neo4j solution

A neo4j db is set up with labels Cigar, House, Nationality, Pet, Drink, Color. They also have these values as property 'thing'.

- Node Labels
    - House (number)
    - Cigar
    - Pet
    - Nationality
    - Color
    - Drink
- Relationships
    - PAIRED
    - NOT_PAIRED
    - SUPER_NEIGHBORS
        - directed, from higher house no to lower. Implies there is a NEIGHBORS between these nodes as well.
    - NEIGHBORS
        - undirected
    - NOT_NEIGHBORS
    
- Repeatable logical connections
    - PAIRS inherit relationships
        - Any PAIREDs inherit *all relationships* to third nodes from the other
    - One and only one of each type per house
        - A node PAIRED with another is NOT_PAIRED with every other node of both types
            - Not necessary to run this check. Nodes of the same kind are all set as NOT_PAIRED at the beginning, and PAIRED nodes will inherit these.
    - SUPER_NEIGHBORS implies NEIGHBORS
    - If a node SUPER_NEIGHBORS a house node, it is paired with the house node of 1+ number

CREATE (node:Cigar {thing:new_node_name, lbl:"cigar"});

//Assign direct hints                   
UNWIND [['English', 'red'],
 ['Swedish', 'dogs'],
 ['Danish', 'tea'],
 ['green', 'coffee'],
 ['Pall Mall', 'birds'],
 ['yellow', 'Dunhill'],
 ['2', 'milk'],
 ['Norwegian', '0'],
 ['Blue Masters', 'beer'],
 ['German', 'Prince'],
 ['Norwegian', 'blue']] AS pairz
WITH pairz
 MATCH (n1), (n2)
 WHERE n1.thing = pairz[0]
     AND n2.thing = pairz[1]

WHERE n1.thing = neighbs[0]
    AND n2.thing = neighbs[1]
WITH n1, n2
MERGE (n1)-[:NEIGHBORS]-(n2);

//Assign directed neighbor hint.
// Green is to the left of white.
MATCH (w:Color {thing:'white'}), (g:Color {thing:'green'})
MERGE (w)-[:SUPER_NEIGHBORS]->(g);
                         
//Set house types to integers for super_neighbor assignment

SET h.thing = toInteger(h.thing);

MATCH (house1:House), (house2:House)
WHERE house1.thing + 1 = house2.thing
MERGE (house2)-[:SUPER_NEIGHBORS]->(house1)
MERGE (house1)-[:NEIGHBORS]-(house2);
               
//every node is not paired to any of its own type: Each house only has one of each kind of thing
MATCH (n1), (n2)
WHERE n1.lbl = n2.lbl
    AND id(n1) <> id(n2)
MERGE (n1)-[:NOT_PAIRED]-(n2);
 
// Green cannot be House no 4, as it has to have white to the right. Likewise, House no 1 cannot be white, As it has to have green to the left (and 0 is taken by norwegian)
// This may end up being redundant with the concept of super_neighbors

/////////////////

// Inferenced relationships from other relationships.
// Any super neighbors are neighbors
MATCH (n1)-[:SUPER_NEIGHBORS]->(n2)
MERGE (n1)-[:NEIGHBORS]-(n2);
                         
//any neighbors are not paired
MATCH (n1)-[:NEIGHBORS]-(n2)
MERGE (n1)-[:NOT_PAIRED]-(n2);
                          
//any paired are not neighbors
MATCH (n1)-[:PAIRED]-(n2)
MERGE (n1)-[:NOT_NEIGHBORS]-(n2);
                      
//////////                          
// Maximum numbers of relationships

// max 1 paired from one node to any other of a given type.
// every node paired with another, is not paired with all the others of the same type.
MATCH (n1)-[:PAIRED]-(n2), (another_n2_type)
WHERE n2.lbl = another_n2_type.lbl
    AND id(n2) <>  id(another_n2_type)
MERGE (n1)-[:NOT_PAIRED]-(another_n2_type);
                         
// max 2 neighbors of a single type from a given node.
// if a node neighbors two which are of the same type, it does not neighbor the remaining 3 of that type
MATCH (n3)-[:NEIGHBORS]-(n1)-[:NEIGHBORS]-(n2)
WHERE id(n3) <> id(n2)
    AND labels(n3) = labels(n2)
WITH n1, n2, n3, labels(n2) as neighbor_label
MATCH (n4)
WHERE id(n4) <> id(n1)
    AND id(n4) <> id(n2)
    AND id(n4) <> id(n3)
    AND labels(n4) = neighbor_label
MERGE (n1)-[:NOT_NEIGHBORS]-(n4);

// Houses 0 and 4 only have one neighbor
MATCH (n1:House)-[:NEIGHBORS]-(n2)

    AND id(n3) <> id(n2)
    AND labels(n3) = neighbor_label
MERGE (n1)-[:NOT_NEIGHBORS]-(n3);

WHERE id(n1) <> id(n3)

WHERE id(n1) <> id(n3)
MERGE (n1)<-[:SUPER_NEIGHBORS]-(n3);

MATCH (n1)-[:PAIRED]-(n2)-[:NOT_NEIGHBORS]-(n3)
WHERE id(n1) <> id(n3)
MERGE (n1)-[:NOT_NEIGHBORS]-(n3);

// If there is only one of a type that isn't paired with a given node, they are paired.
